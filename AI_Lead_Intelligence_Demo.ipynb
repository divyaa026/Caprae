```python
# Create sample lead data (same as your dashboard)
sample_leads = [
    {
        "id": "1",
        "company": {
            "name": "TechFlow Inc",
            "industry": "SaaS",
            "size": "201-500",
            "location": "San Francisco, CA"
        },
        "recent_signals": [
            "Raised $20M Series B - 2 weeks ago",
            "New CTO hired from Google"
        ]
    },
    {
        "id": "2", 
        "company": {
            "name": "DataStream Corp",
            "industry": "Analytics",
            "size": "51-200", 
            "location": "Austin, TX"
        },
        "recent_signals": [
            "40% team expansion this quarter",
            "New VP Engineering from Microsoft"
        ]
    },
    {
        "id": "3",
        "company": {
            "name": "CloudNinja Systems",
            "industry": "DevOps",
            "size": "11-50",
            "location": "Seattle, WA"  
        },
        "recent_signals": [
            "Series A funding announced",
            "Expanding internationally"
        ]
    }
]

# Display leads as DataFrame
leads_df = pd.DataFrame([
    {
        "Company": lead["company"]["name"],
        "Industry": lead["company"]["industry"],
        "Size": lead["company"]["size"],
        "Location": lead["company"]["location"],
        "Key Signals": " | ".join(lead["recent_signals"])
    } for lead in sample_leads
])

print("📋 Sample Lead Data:")
print("=" * 80)
display(leads_df)
```

```python
# AI Lead Analysis Function (mirrors your dashboard's AI service)
def analyze_lead_with_ai(lead_data):
    """
    Analyze a lead using Gemini AI - same logic as your dashboard
    """
    try:
        company = lead_data["company"]
        signals = " | ".join(lead_data.get("recent_signals", []))
        
        prompt = f"""
        Analyze this lead for sales intelligence and respond with ONLY a valid JSON object:
        
        Company: {company["name"]}
        Industry: {company["industry"]}
        Size: {company["size"]}
        Location: {company["location"]}
        Recent Signals: {signals}
        
        Generate realistic sales intelligence in this exact JSON format:
        {{
          "confidence": [number between 65-95],
          "insights": "[2-3 sentence analysis of lead potential and timing]",
          "priority_score": [number between 1-10],
          "key_opportunities": ["opportunity 1", "opportunity 2", "opportunity 3"],
          "recommended_timing": "[timing recommendation]",
          "approach_strategy": "[best approach strategy]"
        }}
        """
        
        print(f"🔄 Analyzing {company['name']} with Gemini AI...")
        start_time = time.time()
        
        response = model.generate_content(prompt)
        analysis_time = time.time() - start_time
        
        # Extract JSON from response
        response_text = response.text.strip()
        json_start = response_text.find('{')
        json_end = response_text.rfind('}') + 1
        
        if json_start >= 0 and json_end > json_start:
            json_text = response_text[json_start:json_end]
            ai_analysis = json.loads(json_text)
            
            # Add metadata
            ai_analysis["analysis_time"] = round(analysis_time, 2)
            ai_analysis["ai_provider"] = "Gemini Pro"
            ai_analysis["timestamp"] = datetime.now().isoformat()
            
            return ai_analysis
        else:
            raise ValueError("No valid JSON found in response")
            
    except Exception as e:
        print(f"❌ AI Analysis failed for {company['name']}: {e}")
        # Fallback to intelligent mock data
        return {
            "confidence": np.random.randint(70, 90),
            "insights": f"High potential lead in {company['industry']} sector with recent growth signals.",
            "priority_score": np.random.randint(6, 9),
            "key_opportunities": ["Market expansion", "Technology upgrade", "Team scaling"],
            "recommended_timing": "Contact within 48 hours",
            "approach_strategy": "Direct outreach with value proposition",
            "analysis_time": 0.0,
            "ai_provider": "Mock (Fallback)",
            "timestamp": datetime.now().isoformat()
        }

# Test with first lead
print("🚀 Testing AI Lead Analysis:")
print("=" * 60)

test_lead = sample_leads[0]
analysis_result = analyze_lead_with_ai(test_lead)

print(f"\n📊 Analysis Results for {test_lead['company']['name']}:")
print("-" * 50)
for key, value in analysis_result.items():
    if key == "key_opportunities":
        print(f"{key.replace('_', ' ').title()}: {', '.join(value)}")
    else:
        print(f"{key.replace('_', ' ').title()}: {value}")
```

## 📧 Section 5: AI Outreach Generation

Let's demonstrate the AI-powered outreach generation feature from your dashboard.

```python
# AI Outreach Generation (mirrors your dashboard feature)
def generate_ai_outreach(lead_data, analysis_data):
    """
    Generate personalized outreach using Gemini AI - same as your dashboard
    """
    try:
        company = lead_data["company"]
        signals = " | ".join(lead_data.get("recent_signals", []))
        
        prompt = f"""
        Create personalized outreach strategy for this sales lead:
        
        Company: {company["name"]}
        Industry: {company["industry"]} 
        Size: {company["size"]}
        Recent Signals: {signals}
        AI Confidence: {analysis_data.get("confidence", 80)}%
        Key Opportunities: {", ".join(analysis_data.get("key_opportunities", []))}
        
        Provide EXACTLY this format:
        
        Email Subject: [compelling subject line, 8 words max]
        
        Talking Points:
        • [key point 1 based on recent signals]
        • [key point 2 about industry challenges]  
        • [key point 3 about competitive advantage]
        • [key point 4 about ROI/implementation]
        
        Timing: [optimal contact timing]
        
        Approach: [recommended strategy based on company profile]
        
        Keep professional but personalized tone.
        """
        
        print(f"📧 Generating outreach strategy for {company['name']}...")
        
        response = model.generate_content(prompt)
        
        # Parse the structured response
        lines = response.text.strip().split('\n')
        parsed_outreach = {
            "email_subject": "",
            "talking_points": [],
            "timing": "",
            "approach": "",
            "ai_provider": "Gemini Pro",
            "generated_at": datetime.now().isoformat()
        }
        
        current_section = None
        for line in lines:
            line = line.strip()
            if line.startswith("Email Subject:"):
                parsed_outreach["email_subject"] = line.replace("Email Subject:", "").strip()
            elif line.startswith("Talking Points:"):
                current_section = "talking_points"
            elif line.startswith("Timing:"):
                parsed_outreach["timing"] = line.replace("Timing:", "").strip()
                current_section = None
            elif line.startswith("Approach:"):
                parsed_outreach["approach"] = line.replace("Approach:", "").strip()
                current_section = None
            elif current_section == "talking_points" and line.startswith("•"):
                parsed_outreach["talking_points"].append(line.replace("•", "").strip())
        
        return parsed_outreach
        
    except Exception as e:
        print(f"❌ Outreach generation failed: {e}")
        # Fallback strategy
        return {
            "email_subject": f"Partnership opportunity with {company['name']}",
            "talking_points": [
                "Reference recent company growth and achievements",
                "Address industry-specific challenges and solutions", 
                "Highlight competitive advantages and success stories",
                "Discuss ROI timeline and implementation approach"
            ],
            "timing": "Contact within 24-48 hours",
            "approach": "Professional outreach with value proposition",
            "ai_provider": "Mock (Fallback)",
            "generated_at": datetime.now().isoformat()
        }

# Generate outreach for analyzed lead
print("🚀 Testing AI Outreach Generation:")
print("=" * 60)

outreach_strategy = generate_ai_outreach(test_lead, analysis_result)

print(f"\n📧 Outreach Strategy for {test_lead['company']['name']}:")
print("-" * 50)
print(f"📝 Email Subject: {outreach_strategy['email_subject']}")
print(f"\n💬 Talking Points:")
for i, point in enumerate(outreach_strategy['talking_points'], 1):
    print(f"   {i}. {point}")
print(f"\n⏰ Timing: {outreach_strategy['timing']}")
print(f"🎯 Approach: {outreach_strategy['approach']}")
print(f"🤖 Generated by: {outreach_strategy['ai_provider']}")
```

```python
# Generate smart recommendations (mirrors dashboard AI recommendations)
def generate_smart_recommendations(analyses):
    """
    Generate strategic recommendations based on all lead analyses
    """
    try:
        # Prepare summary for AI
        lead_summary = []
        for analysis in analyses:
            lead_summary.append({
                "name": analysis["company_name"],
                "confidence": analysis["confidence"],
                "priority": analysis["priority_score"],
                "opportunities": analysis["key_opportunities"]
            })
        
        prompt = f"""
        Analyze these {len(analyses)} sales leads and provide strategic insights for the sales team:
        
        {json.dumps(lead_summary, indent=2)}
        
        Provide actionable sales strategy insights in 2-3 sentences focusing on:
        1. Market opportunities and timing
        2. Prioritization strategy 
        3. Resource allocation recommendations
        
        Keep response conversational and actionable for sales team.
        """
        
        print("🧠 Generating strategic recommendations...")
        response = model.generate_content(prompt)
        
        # Also identify top leads by category
        hot_prospects = sorted(analyses, key=lambda x: x["confidence"], reverse=True)[:2]
        high_priority = sorted(analyses, key=lambda x: x["priority_score"], reverse=True)[:2]
        
        return {
            "ai_insights": response.text.strip(),
            "hot_prospects": [lead["company_name"] for lead in hot_prospects],
            "high_priority": [lead["company_name"] for lead in high_priority],
            "generated_at": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        }
        
    except Exception as e:
        print(f"❌ Recommendations generation failed: {e}")
        return {
            "ai_insights": "Focus on leads with recent funding events and strong growth signals. Prioritize warm connections and time-sensitive opportunities.",
            "hot_prospects": [analyses[0]["company_name"], analyses[1]["company_name"]], 
            "high_priority": [analyses[0]["company_name"], analyses[1]["company_name"]],
            "generated_at": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        }

# Generate recommendations
recommendations = generate_smart_recommendations(all_lead_analyses)

print("\n🎯 AI-Generated Strategic Recommendations:")
print("=" * 70)
print(f"📝 Strategic Insights:")
print(f"   {recommendations['ai_insights']}")
print(f"\n🔥 Hot Prospects: {', '.join(recommendations['hot_prospects'])}")
print(f"⭐ High Priority: {', '.join(recommendations['high_priority'])}")
print(f"🕒 Generated: {recommendations['generated_at']}")
```

```python
# Performance testing and rate limiting demonstration
def test_ai_performance():
    """
    Test AI performance metrics and rate limiting behavior
    """
    print("⚡ Testing AI Performance & Rate Limits:")
    print("=" * 60)
    
    # Track performance metrics
    response_times = []
    success_count = 0
    error_count = 0
    
    test_prompts = [
        "Analyze TechFlow Inc for sales potential",
        "Generate confidence score for DataStream Corp", 
        "Evaluate CloudNinja Systems lead quality",
        "Provide strategic insights for SaaS leads",
        "Recommend outreach timing for tech companies"
    ]
    
    print("🔄 Running 5 consecutive AI requests...")
    
    for i, prompt in enumerate(test_prompts, 1):
        try:
            start_time = time.time()
            response = model.generate_content(prompt)
            end_time = time.time()
            
            response_time = end_time - start_time
            response_times.append(response_time)
            success_count += 1
            
            print(f"   Request {i}: ✅ {response_time:.2f}s")
            
            # Add delay to be respectful of rate limits
            time.sleep(0.5)
            
        except Exception as e:
            error_count += 1
            print(f"   Request {i}: ❌ {str(e)[:50]}...")
    
    # Calculate performance metrics
    if response_times:
        avg_time = np.mean(response_times)
        min_time = np.min(response_times)
        max_time = np.max(response_times)
        
        print(f"\n📊 Performance Summary:")
        print(f"   Success Rate: {success_count}/{len(test_prompts)} ({success_count/len(test_prompts)*100:.1f}%)")
        print(f"   Average Response Time: {avg_time:.2f}s")
        print(f"   Fastest Response: {min_time:.2f}s")
        print(f"   Slowest Response: {max_time:.2f}s")
        
        # Rate limit info
        print(f"\n🔒 Gemini Free Tier Limits:")
        print(f"   Rate Limit: 60 requests per minute")
        print(f"   Current Test: {len(test_prompts)} requests in {sum(response_times):.1f}s")
        print(f"   Recommended Delay: 1-2s between requests")
        
    return {
        "success_rate": success_count/len(test_prompts)*100,
        "avg_response_time": avg_time if response_times else 0,
        "total_requests": len(test_prompts),
        "errors": error_count
    }

# Run performance test
performance_results = test_ai_performance()

# Error handling demonstration
print(f"\n🛡️ Error Handling Test:")
print("-" * 40)

def test_error_handling():
    """Test graceful error handling"""
    try:
        # Test with invalid/empty prompt
        response = model.generate_content("")
        print("✅ Empty prompt handled successfully")
    except Exception as e:
        print(f"❌ Empty prompt error (expected): {str(e)[:50]}...")
        
    try:
        # Test with very long prompt (might hit limits)
        long_prompt = "Analyze this company: " + "data " * 1000
        response = model.generate_content(long_prompt)
        print("✅ Long prompt handled successfully") 
    except Exception as e:
        print(f"❌ Long prompt error (expected): {str(e)[:50]}...")
        
    print("✅ Error handling working correctly - falls back gracefully")

test_error_handling()
```
